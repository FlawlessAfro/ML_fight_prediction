```python
df_original=pd.read_csv('../bronze/ufc-master.csv')
df_odds = df_original.copy()
#df_odds= df_odds[['R_fighter', 'B_fighter','R_odds','B_odds','date']]

df_odds['date'] = df_odds['date'].apply(lambda x: pd.to_datetime(x).strftime('%m-%d-%Y'))

df_odds
```

```python
red_win = df_odds.loc[df_odds['Winner'] == 'Red']
red_win['Winner'] = red_win['R_fighter']
blue_win = df_odds.loc[df_odds['Winner'] == 'Blue']
blue_win['Winner'] = blue_win['B_fighter']
df_winner = pd.concat([red_win,blue_win])
```

```python
win_counts = df_odds.groupby(['Winner', 'B_Stance', 'R_Stance']).size().reset_index(name='count')
win_counts
```

```python
blue_win_counts = df_odds[df_odds['Winner'] == 'Blue'].groupby(['B_Stance']).size().reset_index(name='Blue_win_count')
blue_win_counts
```

```python
class_stats = df_odds.groupby('weight_class').agg(
    count=('weight_class', 'count'),
    avg_fight_time=('total_fight_time_secs', 'mean')
).reset_index()
class_stats= class_stats.sort_values(by='avg_fight_time', ascending=False)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(class_stats['weight_class'], class_stats['avg_fight_time'], color='skyblue')

# Add labels and title
plt.xlabel('Weight Class')
plt.ylabel('Avg Fight Time (seconds)')
plt.title('Average Fight Time by Weight Class')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

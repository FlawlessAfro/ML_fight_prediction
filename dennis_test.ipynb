```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
import numpy as np
```

```python
df_original = pd.read_csv(r'bronze\ufc-master.csv')
df_copy = df_original.copy()
```

```python
Nan_Values = df_copy.isna().sum()
```

OHE B_Stance og R_Stance

```python
df_copy = df_copy.dropna(subset=['B_Stance'])
```

```python
df_copy = df_copy.dropna(subset=['R_Stance'])
```

```python
df_copy['B_Stance'].isna().sum()
```

```python
ohe_columns = ['B_Stance', 'R_Stance']
ohe_stance = OneHotEncoder(sparse_output=False)
ohe_stance.fit(df_copy[ohe_columns])
ohe_column_name = ohe_stance.get_feature_names_out(ohe_columns)
df_copy[ohe_column_name] = ohe_stance.transform(df_copy[ohe_columns])
```

```python
#DOBBELTSJEKK OM DU FIKK 2 KOLONNER MED NAVN: B_Stance_Switch
df_copy.drop(columns=['B_Stance_Switch '], inplace=True)
```

DROPPER WEIGHTCLASS RANK

```python
columns_to_drop = []
for column in df_copy.columns:
    if 'weight_rank' in column:
        columns_to_drop.append(column)

print(columns_to_drop)
```

```python
df_copy.drop(columns=columns_to_drop, inplace=True)
```

```python
df_copy
```

Endrer datoformat

```python
df_copy['date'] = df_copy['date'].apply(lambda x: pd.to_datetime(x).strftime('%d-%m-%Y'))
```

OHE Winner-column

```python
ohe_columns = ['Winner']
ohe_stance = OneHotEncoder(sparse_output=False)
ohe_stance.fit(df_copy[ohe_columns])
ohe_column_name = ohe_stance.get_feature_names_out(ohe_columns)
df_copy[ohe_column_name] = ohe_stance.transform(df_copy[ohe_columns])
```

OHE Better_rank-column

```python
ohe_columns = ['better_rank']
ohe_stance = OneHotEncoder(sparse_output=False)
ohe_stance.fit(df_copy[ohe_columns])
ohe_column_name = ohe_stance.get_feature_names_out(ohe_columns)
df_copy[ohe_column_name] = ohe_stance.transform(df_copy[ohe_columns])
```

Fyller ut total_fight_time_secs

```python
average_time = df_copy['total_fight_time_secs'].mean()
```

```python
def update_time(row):
    if pd.isna(row['total_fight_time_secs']):
        if isinstance(row['finish'], str) and 'dec' in row['finish']:
            return 900
        else:
            return average_time
    else:
        return row['total_fight_time_secs']
```

```python
df_copy['total_fight_time_secs'] = df_copy.apply(update_time, axis=1)
```

ENDRER ODDS TIL PROSENT

```python
columns_odds_to_percent = []
for column in df_copy.columns:
    if 'odd' in column:
        columns_odds_to_percent.append(column)

print(columns_odds_to_percent)
```

```python
def odds_to_percent_function(odds):
    
    if odds > 0:
        return (100/(odds + 100))
    if odds < 0:
        odds = abs(odds)
        return ((odds)/(odds + 100))
```

```python
for column in columns_odds_to_percent:
    df_copy[column] = df_copy[column].apply(odds_to_percent_function)
```

```python
df_copy.to_csv('df_ufc_masters.csv', index=False)
```

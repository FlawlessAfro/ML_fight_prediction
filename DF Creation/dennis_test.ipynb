```python
average_time = df_copy['total_fight_time_secs'].mean()
```

```python
def update_time(row):
    if pd.isna(row['total_fight_time_secs']):
        if isinstance(row['finish'], str) and 'dec' in row['finish']:
            return 900
        else:
            return average_time
    else:
        return row['total_fight_time_secs']
```

```python
df_copy['total_fight_time_secs'] = df_copy.apply(update_time, axis=1)
```

ENDRER ODDS TIL PROSENT

```python
columns_odds_to_percent = []
for column in df_copy.columns:
    if 'odd' in column:
        columns_odds_to_percent.append(column)

print(columns_odds_to_percent)
```

```python
def odds_to_percent_function(odds):
    
    if odds > 0:
        return (100/(odds + 100))
    if odds < 0:
        odds = abs(odds)
        return ((odds)/(odds + 100))
```

```python
for column in columns_odds_to_percent:
    df_copy[column] = df_copy[column].apply(odds_to_percent_function)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, accuracy_score

# Re-loading the dataset
latest_file_path = '/mnt/data/df_ufc_masters_w_reversed.csv'
df_latest = pd.read_csv(latest_file_path)

# Specified features for the model
feature_names = [
    'B_avg_SIG_STR_landed',
    'B_avg_SIG_STR_pct',
    'B_avg_SUB_ATT',
    'B_avg_TD_landed',
    'B_avg_TD_pct',
    'R_avg_SIG_STR_landed',
    'R_avg_SIG_STR_pct',
    'R_avg_SUB_ATT',
    'R_avg_TD_landed',
    'R_avg_TD_pct',
    'B_Height_cms',
    'B_Reach_cms',
    'R_Height_cms',
    'R_Reach_cms'
]

# Extracting features and target variable
X = df_latest[feature_names]
y = df_latest['Winner']

# Encoding the target variable (assuming 'Red' as 1 and 'Blue' as 0)
y_encoded = y.apply(lambda x: 1 if x == 'Red' else 0)

# Applying standard scaling to the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Splitting the data into training and test sets
X_train_scaled, X_test_scaled, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.2, random_state=42)

# Creating and training a logistic regression model on the scaled data
log_reg_model = LogisticRegression()
log_reg_model.fit(X_train_scaled, y_train)

# Predicting probabilities on the test set
y_pred_proba = log_reg_model.predict_proba(X_test_scaled)

# Converting probabilities to percentages
y_pred_percent = y_pred_proba * 100

# Display the first few predictions as percentages
y_pred_percent[:5]
```

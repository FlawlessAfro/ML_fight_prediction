```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
df =  pd.read_csv(r'C:\Users\DY\Documents\GitHub\UFCproject\CSV Files\df_ufc_masters.csv')
```

```python
selected_features = [
    'R_fighter',
    'R_avg_SIG_STR_landed',
    'R_avg_SIG_STR_pct',
    'R_avg_SUB_ATT',
    'R_avg_TD_landed',
    'R_avg_TD_pct',
    'R_Height_cms',
    'R_Reach_cms',
    'R_age'
]
```

```python
df_selected = df[selected_features]
```

```python
fighter_counts = df_selected['R_fighter'].value_counts().rename('count')
```

```python
df_selected = df_selected.merge(fighter_counts, left_on='R_fighter', right_index=True)
```

```python
grouped_features = [
    'R_avg_SIG_STR_landed',
    'R_avg_SIG_STR_pct',
    'R_avg_SUB_ATT',
    'R_avg_TD_landed',
    'R_avg_TD_pct'
]
df_grouped = df_selected.groupby('R_fighter')[grouped_features].mean()
```

```python
df_final = df_grouped.merge(df_selected[['R_fighter', 'count', 'R_Height_cms', 'R_Reach_cms', 'R_age']].drop_duplicates(), on='R_fighter')
```

```python
df_final
```

```python
df_final_unique = df_final.drop_duplicates(subset='R_fighter')
```

```python
# List of columns to apply np.log
columns_to_log = ['R_avg_SIG_STR_landed', 'R_avg_SUB_ATT', 'R_avg_TD_landed']

# Applying np.log to each column in the list
for col in columns_to_log:
    # Adding a small constant to avoid log(0) which is undefined
    df_final_unique[col] = np.log(df[col] + 1e-5)
```

```python
# NÅ SKAL DU STANDARDSCALE
features_std_sc = [
    'R_age',
    'R_avg_SIG_STR_landed',
    'R_avg_SIG_STR_pct',
    'R_avg_SUB_ATT',
    'R_avg_TD_landed',
    'R_avg_TD_pct',
    'R_Height_cms',
    'R_Reach_cms'
]
```

```python
df_final_unique[features_std_sc].isna().sum()
```

```python
sc = StandardScaler()
df_final_unique[[
    'R_age_ss',
    'R_avg_SIG_STR_landed_ss',
    'R_avg_SIG_STR_pct_ss',
    'R_avg_SUB_ATT_ss',
    'R_avg_TD_landed_ss',
    'R_avg_TD_pct_ss',
    'R_Height_cms_ss',
    'R_Reach_cms_ss']] = sc.fit_transform(df_final_unique[features_std_sc])
```

```python
columns_to_drop = [
    'R_age',
    'R_avg_SIG_STR_landed',
    'R_avg_SIG_STR_pct',
    'R_avg_SUB_ATT',
    'R_avg_TD_landed',
    'R_avg_TD_pct',
    'R_Height_cms',
    'R_Reach_cms'
]

df_final_unique = df_final_unique.drop(columns=columns_to_drop)
```

```python
rename_dict = {
    'R_age_ss': 'R_age',
    'R_avg_SIG_STR_landed_ss': 'R_avg_SIG_STR_landed',
    'R_avg_SIG_STR_pct_ss': 'R_avg_SIG_STR_pct',
    'R_avg_SUB_ATT_ss': 'R_avg_SUB_ATT',
    'R_avg_TD_landed_ss': 'R_avg_TD_landed',
    'R_avg_TD_pct_ss': 'R_avg_TD_pct',
    'R_Height_cms_ss': 'R_Height_cms',
    'R_Reach_cms_ss': 'R_Reach_cms'
}

df_final_unique = df_final_unique.rename(columns=rename_dict)
```

```python
df_final_unique.to_csv('unike_fighters.csv', index=False)
```
